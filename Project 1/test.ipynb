```python
import plotly.express as px
import matplotlib.pyplot as plt

x1 = [39000, 35000, 30000, 22000, 14000, 7000, 0]
y1 = [0, 110000, 180000, 260000, 310000, 340000, 350000]

x2 = [38000, 32000, 26000, 17000, 10000, 5000, 0]
y2 = [0, 20000, 60000, 120000, 180000, 225000, 250000]
```

```python
plt.figure(figsize=(15,10))
plt.plot(x1, y1, marker='o', color='g', label='Normal')
plt.plot(x2, y2, marker='o', color='r', label='Drought')
plt.title("Normal vs Drought")
plt.legend()
plt.xlabel("Digital Watches")
plt.ylabel("Pounds of Wheat")
plt.savefig("macro.png", bbox_inches="tight")
```
